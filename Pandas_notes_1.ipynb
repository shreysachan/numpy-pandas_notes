```python
df[df['Sunshine']>3]
```

```python
# if we want to print only the  sunshine column where the we get only the values of those entries which are more 
# than a 'x' amount:

df['Sunshine'][df['Sunshine']>3]
```

now if we just want those rows where we want some certain other column entries for which the corresponding Sunshine value is greater than 3, we use:

```python
#condition 1 (where we want criteria for rainfall columns to be sunshine > 4)
df['Rainfall'][df['Sunshine']>4]
```

```python
#condition 2 (where we want criteria for rainfall columns to be sunshine is max())
df['Rainfall'][df['Sunshine']==df['Sunshine'].max()]
```

```python
# we can also have multiple of these apart from rainfall at once:

df[['Rainfall','Evaporation']][df['Sunshine']==df['Sunshine'].max()]
```
