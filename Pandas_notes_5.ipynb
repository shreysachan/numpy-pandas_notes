Handle Missing Data: replace function

```python
import pandas as pd
import numpy as np
df = pd.read_csv("new_weather_data.csv")
df.set_index("day")
df
```

```python
#basic workings of replace method, here we specify the value to be replaced in the first parameter and the value to 
#be replaced with as the second parameter
df1 = df.replace(-99999,np.NaN)
df1
```

```python
#we can also provide a list in the previous step
df1 = df.replace([-99999,7],np.NaN)
df1
```

```python
#to make the previous step more precise we can use a dictionary instead of a list in the previous step instead
df2= df.replace({
    'temperature':-99999,
    'windspeed': [-99999,7],

    'event':'0'
}, np.nan)
df2
```

```python
#another way of using a dictionary in the previous step
df3 = df.replace({
    -99999:np.nan,
    '0':'hulabuga'
})
df3
```

```python
#to replace a certain pattern of strings or numbers from the ds we use regex
df4 = df.replace('[A-za-z]','', regex=True)
df4
```

```python
#the previous method will replace all the values in the event column as well, to counter that:
df5= df.replace({
    'temperature':'[A-za-z]',
    'windspeed': '[A-za-z]'
    },'',regex=True)
df5
```

```python
#Replacing list with another list (here if we want to replace a certain string values with their numeric counterparts)

stu_data = {'score':['exceptional','average', 'good', 'poor', 'average', 'exceptional']
           , 'name':['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']}
df6 = pd.DataFrame(stu_data)
df7 = df6.replace(['poor', 'average', 'good', 'exceptional'],[0,1,2,3])
df7
```
