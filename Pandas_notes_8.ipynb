Merge Dataframes (Visualize it as a venn diagram)

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city':['ny', 'chicago','orlando'],
    'temperature':[21,12,23]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':[ 'chicago','ny','orlando'],
    'humidity':[1,2,3]
})
df2
```

```python
#used to merge two dataframes based "on" what column
df3 = pd.merge(df1,df2, on="city")
df3
```

```python
df4 = pd.DataFrame({
    'city':['ny', 'chicago','orlando', 'baltimore'],
    'temperature':[21,12,23,33]
})
df4
#here we have a df that has 5 entries in the joining column, with only 2 common entries
```

```python
df5 = pd.DataFrame({
    'city':[ 'chicago','ny','sanfran'],
    'humidity':[1,2,4]
})
df5
#here we have a df that has only 3 entries in the joining column, with 2 common entries
```

```python
df6 = pd.merge(df4,df5)
df6
#when we join these two df based on the city column, we get a df that only shows the common entries
#this is because by default it performs "inner" join in the how section without the need to specify 
```

```python
df7 = pd.merge(df4,df5, how="outer", indicator=True) #we can also specify how="left/right" to get only the 
# left join or right join
df7
#when we explicitly specify how="outer" then we get the desired result 
# indicator is used to specify from what side of the dataset the data is taken from 
```

```python
#here we have a dataframe with some certain values of new york and chicago
dfa = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
dfa
```

```python
#here we have a dataset with different values for newyork and chicago compared to the prev. df
dfb = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
dfb
```

```python
#if we merge the prev. 2 df together it will create 4 separate columns instead of 2 columns using suffixes
dfc= pd.merge(dfa,dfb,on="city",how="outer", suffixes=('_first','_second'))
dfc
#it will right 'humidity_x' and 'humdity_y' if we do not specify suffixes
```
