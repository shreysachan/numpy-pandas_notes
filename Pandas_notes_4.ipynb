Handle Missing Data: fillna, dropna, interpolate

```python
import pandas as pd 
df = pd.read_csv("weather_data.csv", parse_dates=["day"]) #parse dates is used to parse the values in a 
#column as date-time objects
```

```python
df.columns
df.set_index("day", inplace=True) #remember, without inplace df will not be updated(not permanent)
```

```python
df
```

```python
new_df= df.fillna(0) #fills all the NaN values with 0
new_df
```

```python
#fills all the NaN values according to the dictionary
new_df3 = df.fillna(
    {
        "temperature": 0, "windspeed":0, "event":"no event"
    }
)
new_df3
```

```python
#used to forward fill the values from the previous cell 

new_df2= df.ffill()
new_df2
```

```python
#used to backward fill the values from the previous cell 
#we can also change the axis, in what direction we want to fill the values
new_df4=df.bfill(axis="columns")
new_df4
```

```python
new_df4=df.bfill(axis="columns", limit=1)
#we can also decide if we want only (N) values to be filled backwards/forwards
new_df4
```

```python
#used to find better values for filling missing values using linear interpolation
new_df5= df.interpolate(method="time")
#method="time" also considers the gap of time and not just gap of values for interpolation
new_df5 
```

```python
# this is used to drop all the rows with null values if there is at least one na value in row
new_df6 = df.dropna()
new_df6
```

```python
# this is used when we want to drop the row, only when the entire row contains na values
new_df7 = df.dropna(how="all")
new_df7
```

```python
#specifies the minimum number of non-null values required for a row or column to be retained.
df.dropna(thresh=2)
df
```
